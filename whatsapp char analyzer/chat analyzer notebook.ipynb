```python
import numpy as np
import pandas as pd
import seaborn as sns
import re
import datetime as dt
import matplotlib.pyplot as plt
```

```python
# importing chat file
f = open('chat.txt','r',encoding='utf-8')
data = f.read()
print(data)
```

```python
# extracting the date and time portion using regular expression

pattern = '\d{1,2}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{1,2}\s[APMapm]{2}\s-\s' 
```

```python
# splitting on the extracting pattern to seperate messages
messages =  re.split(pattern,data)[1:]
print(messages)
```

```python
# extracing dateTime

dates = re.findall(pattern,data)
dates
```

```python
# converting extracted messages and dateTime into data frame

df=pd.DataFrame({'user_messages':messages,'date':dates})
df.head(2)
```

```python
df.info()
```

```python
# converting date column into datetime format 

format_str = '%m/%d/%y, %I:%M\u202f%p - '

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'], format=format_str, errors='coerce')
```

```python
df.head()
```

```python
df.info()
```

```python
# extracting users from the messages 


df = df[df['user_messages'].str.contains(':')]

# Extract names using colon as delimiter
df['users'] = df['user_messages'].str.split(':', n=1).str[0]

```

```python
df
```

```python
# after seperting names from the message no extracting the actual message

df['messages']=df['user_messages'].str.split(':').str[1].str.strip()
```

```python
df.sample()
```

```python
# creating new coumns from date column for later visualization and activities

df['year']=df['date'].dt.year
df['month']=df['date'].dt.month_name()
df['day']=df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df['day_of_week'] = df['date'].dt.day_name()
```

```python
df.head()
```

```python
# dropping user_message and date columns we have extracted all the necessary information it is of no use

df=df.drop(['user_messages','date'],axis=1)
```

Extracting Emojis from the chat

```python
import re

# Define a regular expression pattern to match emojis
emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # Emoticons
                           u"\U0001F300-\U0001F5FF"  # Symbols & Pictographs
                           u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
                           u"\U0001F1E0-\U0001F1FF"  # Flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)

# Extract emojis from messages column
emojis = []
for msg in df['messages']:
    if isinstance(msg, str):  # Check if the value is a string
        emojis.extend(emoji_pattern.findall(msg))
```

```python
emojis
```

Count of the emoji sent by users

```python
from collections import Counter



# Initialize an empty list to store individual emojis
individual_emojis = []

# Split each string of emojis into individual emojis
for emoji in emojis:
    individual_emojis.extend(list(emoji))

# Count the occurrences of each emoji
emoji_counts = Counter(individual_emojis).most_common()

# converting emoji counts into DataFrame

df_emoji = pd.DataFrame(emoji_counts, columns=['Emoji', 'Count'])
```

## Monthly activity of group 

```python
# finding in which month has the most messages
timeline = df.groupby(['year', 'month']).count()['messages'].reset_index()

timeline
```

```python
# concatinating month name with year to get this format (December-2023) which is easily understandable
time = []
for i in range(timeline.shape[0]):
    time.append(timeline['month'][i]+ "-"+str(timeline['year'][i]) )
```

```python
# we have created new column in the timeline dataframe called time
timeline['time'] = time
timeline
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming timeline is your DataFrame
plt.figure(figsize=(10, 6))
sns.barplot(x='time', y='messages', data=timeline)
plt.xlabel('Time')
plt.ylabel('Number of Messages')
plt.title('Timeline of Messages')
plt.xticks(rotation=45)
plt.show()
```
